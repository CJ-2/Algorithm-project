```python
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#table A
plt.figure(figsize=(20, 10))
Size = [50,100,150,200]
kruskal_before = [0.00380857785542806,0.004401286443074544,0.00585635503133138,0.006313800811767578]
kruskal_after = [0.0032939910888671875,0.0042527516682942705,0.005573272705078125,0.00538333257039388]
plt.plot(Size, kruskal_after, marker='o')
plt.plot(Size, kruskal_before, marker='o')
plt.xlabel('Size')
plt.ylabel('Seconds')
plt.title('Table A')
plt.legend(['kruskal_after', 'Kruskal_before'])
```
